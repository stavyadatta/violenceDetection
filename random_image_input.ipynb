```python
def putting_images(label, size):
    j = 0
    while j < size:
        img_num = randint(0, 20000)
        print(f'fight_image_dataset/{label}/img{img_num}.jpg')
        image = cv2.imread(f'fight_image_dataset/{label}/img{img_num}.jpg')
        cv2.imwrite("medium_fight_dataset/{}/img{}.jpg".format(label, j), image)
        j +=1
```
