```python
import os
import cv2
```

```python
no_count = 0
fight_count = 0
for filename in os.listdir('HockeyFights'):
    vid_obj = cv2.VideoCapture('HockeyFights/'+filename)
    while True:
        success, image = vid_obj.read()
        if success:
            cv2.resize(image, (160, 160))
            if "fi" in filename:
                cv2.imwrite("fight_image_dataset/fight/img{}.jpg".format(fight_count), image)
                print("fight_image_dataset/fight/img{}.jpg".format(fight_count))
                fight_count+=1
            else:
                cv2.imwrite("fight_image_dataset/no_fight/img{}.jpg".format(no_count), image)
                print("fight_image_dataset/no_fight/img{}.jpg".format(no_count))
                no_count += 1
        else:
            break
    vid_obj.release() 
    cv2.destroyAllWindows()
```

```python
from __future__ import absolute_import
from __future__  import division
from __future__ import print_function
import tensorflow as tf
import numpy as np
import numpy as np
import os
from PIL import Image
from io import BytesIO
import time
from tensorflow.keras import layers, models, losses, metrics, optimizers
```

```python
num_classes = 2 
input_shape = (160, 160, 3)
np.random.seed(1234)
vg19 = tf.keras.applications.vgg19.VGG19
base_model = vg19(include_top=False, weights='imagenet', input_shape=(160, 160,3))
```

```python
cnn = models.Sequential()
cnn.add(base_model)
cnn.add(layers.Flatten())
model = models.Sequential()
model.add(layers.TimeDistributed(cnn,  input_shape=(30, 160, 160, 3)))
model.add(layers.LSTM(30 , return_sequences= True, input_shape=(160, 160, 3)))
model.add(layers.TimeDistributed(layers.Dense(90)))
model.add(layers.Dropout(0.1))
model.add(layers.GlobalAveragePooling1D())
model.add(layers.Dense(512, activation='relu'))
model.add(layers.Dropout(0.3))
model.add(layers.Dense(num_classes, activation="sigmoid"))
adam = optimizers.Adam(lr=0.0005, beta_1=0.9, beta_2=0.999, epsilon=1e-08)
#model.load_weights('mamonbest947oscombo.hdfs')
rms = optimizers.RMSprop()
model.compile(loss='binary_crossentropy', optimizer=adam, metrics=["accuracy"])
```

```python
from imutils import paths
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.applications.vgg19 import preprocess_input
import os
import cv2
imagePaths = list(paths.list_images('fight_image_dataset'))
```

```python
data1 = []
labels1 = []
i = 0
while i < 100:
    label = imagePaths[i].split(os.path.sep)[-2]
    image = load_img(imagePaths[i], target_size=(160,160))
    image = img_to_array(image)
    image = image.reshape((1, image.shape[0], image.shape[1], image.shape[2]))
    image = preprocess_input(image)
    print(image.shape)
    data1.append(image)
    labels1.append(label)
    i+= 10

    
    
```

```python
data = data1
labels = labels1
from sklearn.preprocessing import LabelBinarizer
from tensorflow.keras.utils import to_categorical
lb = LabelBinarizer()
labels = lb.fit_transform(labels)
labels = to_categorical(labels)

```

```python
from sklearn.model_selection import train_test_split
data = np.array(data)
labels = np.array(labels)
print(labels.shape)
(trainX, testX, trainY, testY) = train_test_split(data, labels,
	test_size=0.20, stratify=labels, random_state=42)
print(data.shape)
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
aug = ImageDataGenerator(
	rotation_range=20,
	zoom_range=0.15,
	width_shift_range=0.2,
	height_shift_range=0.2,
	shear_range=0.15,
	horizontal_flip=True,
	fill_mode="nearest")
```

```python
BS = 2
EPOCHS = 20
H = model.fit(
    trainX, trainY,
	steps_per_epoch=len(trainX) // BS,
	validation_data=(testX, testY),
	validation_steps=len(testX) // BS,
	epochs=EPOCHS
)
```
